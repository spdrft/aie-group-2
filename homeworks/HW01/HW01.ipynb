```python
import numpy as np
```

Массивы

```python
#создадим одномерный массив
a1 = np.array([1, 2, 3])
a1
```

```python
#создадим двумерный массив 
a2 = np.array([['a','b','c'], [1,2,3]])
a2
```

```python
# двумерный массив из нулей, размером 4 на 4
b=np.zeros((4,4)) 
b
```

```python
#создать массив с заданной последовательностью и шагом
c=np.arange(20, 30, 2)
c
```

Демонстрация атрибутов:
вывод формы (shape) и типа данных (dtype) для одного из массивов.

```python
#выведем форму и тип данных для созданного ранее массива a2
print(np.shape(a2))
print(type(a2))
```

Примеры индексирования и срезов:
получение отдельного элемента;
извлечение подмассива с помощью среза.

```python
#элемент массива a2 [0,0] - первый элемент первой строки
a2[0,0]
```

```python
#извлечение подмассива с помощью среза
#извлекаем из всех строк элементы с 0 по 2(не включая)
a2[:, 0:2]
```

```python
#извлекаем первый элемент первой строки
a2[0, 0]
```

Примеры векторных операций и базовых статистик:
поэлементное сложение или умножение массивов;
вычисление средней величины (mean) или другой простой статистики.

```python
#создадим массивы с рандомными целыми числовыми значениями
z1 = np.random.randint(0, 100, 3)
z2 = np.random.randint(0, 100, 3)
print(z1)
print(z2)
```

```python
#поэлементарное сложение
res = z1 + z2
res
```

```python
#умножение массивов
res2 = z1 * z2
res2
```

```python
#вычисление средней величины
print(np.mean(res2))
```
